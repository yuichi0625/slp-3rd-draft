## 12.1
Draw tree structures for the following ATIS phrases:
1. Dallas
1. from Denver
1. after five p.m.
1. arriving in Washington
1. early flights
1. all redeye flights
1. on Thursday
1. a one-way fare
1. any delays in Denver

```python
# 1.Dallas
Tree.fromstring('(NP (ProperNoun Dallas))')
```

```python
# 2.from Denver
Tree.fromstring('(PP (Preposition from) (NP (ProperNoun Denver)))')
```

```python
# 3.after five p.m.
Tree.fromstring('(PP (Preposition after) (NP (CardinalNumber five) (Nominal (Noun p.m.))))')
```

```python
# 4.arriving in Washington
Tree.fromstring('(GerundVP (GerundV arriving) (PP (Preposition in) (NP (ProperNoun Washington))))')
```

```python
# 5.early flights
Tree.fromstring('(NP (AP (Adjective early)) (Nominal (Noun flights)))')
```

```python
# 6.all redeye flights
Tree.fromstring('(NP (PreDet all) (Nominal (Nominal (Noun redeye)) (Noun flights)))')
```

```python
# 7.on Thursday
Tree.fromstring('(PP (Preposition on) (NP (ProperNoun Thursday)))')
```

```python
# 8.a one-way fare
Tree.fromstring('(NP (Det a) (AP (Adjective one-way)) (Nominal (Noun fare)))')
```

```python
# 9.any delays in Denver
Tree.fromstring('(NP (Det any) (Nominal (Nominal (Noun delays)) (PP (Preposition in) (NP (ProperNoun Denver)))))')
```

## 12.2
Draw tree structures for the following ATIS sentences:
1. Does American Airlines have a flight between five a.m. and six a.m.?
1. I would like to fly on American Airlines.
1. Please repeat that.
1. Does American 487 have a first-class section?
1. I need to fly between Philadelphia and Atlanta.
1. What is the fare from Atlanta to Denver?
1. Is there an American Airlines flight from Philadelphia to Dallas?

```python
# 1.Does American Airlines have a flight between five a.m. and six a.m.?
Tree.fromstring("""
(S (Aux Does)
   (NP (ProperNoun American-Airlines))
   (VP (Verb have)
       (NP (Det a)
           (Nominal (Noun flight)))
       (PP (Preposition between)
           (NP (NP (CardinalNumber five)
                   (Nominal (Noun a.m.)))
               and
               (NP (CardinalNumber six)
                   (Nominal (Noun a.m.)))))))
""")
```

```python
# 2.I would like to fly on American Airlines.
Tree.fromstring("""
(S (NP (Pronoun I))
   (VP (Verb would-like)
       (VP (Verb to-fly)
           (PP (Preposition on)
               (NP (ProperNoun American-Airlines))))))
""")
```

```python
# 3.Please repeat that.
```

```python
# 4.Does American 487 have a first-class section?
Tree.fromstring("""
(S (Aux Does)
   (NP (ProperNoun American-487))
   (VP (Verb have)
       (NP (Det a)
           (AP (Adjective first-class))
           (Nominal (Noun section)))))
""")
```

```python
# 5.I need to fly between Philadelphia and Atlanta.
Tree.fromstring("""
(S (NP (Pronoun I))
   (VP (Verb need)
       (VP (Verb to-fly)
           (PP (Preposition between)
               (NP (NP (ProperNoun Philadelphia))
                   and
                   (NP (ProperNoun Atlanta)))))))
""")
```

```python
# 6.What is the fare from Atlanta to Denver?
Tree.fromstring("""
(S (Wh-NP What)
   (VP (Verb is)
       (NP (Det the)
           (Nominal (Nominal (Nominal (Noun fare))
                             (PP (Preposition from)
                                 (NP (ProperNoun Atlanta))))
                    (PP (Preposition to)
                        (NP (ProperNoun Denver)))))))
""")
```
